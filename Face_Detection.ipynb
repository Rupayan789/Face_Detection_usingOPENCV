```
plt.imshow(img)
plt.axis("off")
plt.show()
```

```
import cv2
model = cv2.CascadeClassifier('pattern.xml')
```

```
input = img
output = model.detectMultiScale(input,minSize=(50,50))

print(output)
for face in output:
    x,y,w,h = face
    green = (255,255,0)
    cv2.rectangle(img,(x,y),(x+w,y+h),green,5)

plt.imshow(img)
plt.axis("off")
plt.show()
```
